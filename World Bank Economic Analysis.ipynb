```python
import pandas as pd

file_path = "E:/My Folder/UTD/Maven Analytics/Python/World+Economic+Indicators/WorldBank.xlsx"

df = pd.read_excel(file_path)

df.head()
```

```python
df["Population (M)"] = (df["GDP (USD)"]/df["GDP per capita (USD)"])/1000000
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
file_path = "E:/My Folder/UTD/Maven Analytics/Python/World+Economic+Indicators/HDI.csv"

hdi_df = pd.read_csv(file_path)

hdi_df.head()
```

```python
wld_hdi_2014 = df.query("Year == 2014").merge(hdi_df[["iso3","hdi_2014"]], how='left', left_on='Country Code', right_on="iso3")
wld_hdi_2014
```

```python
wld_hdi_2014.describe()
```

```python
gdp_pivot = pd.pivot_table(
    df,
    values=['GDP (USD)'],      # Columns to aggregate
    index=['Year'],            # Columns to use as rows
    columns=['Region'],        # Columns to use as columns
    aggfunc='sum',             # Aggregation function
    fill_value=0               # Fill missing values with 0
)

gdp_pivot
```

```python
pop_pivot = pd.pivot_table(
    df,
    values=['Population (M)'],      # Columns to aggregate
    index=['Year'],             # Columns to use as rows
    columns=['Region'],                  # Columns to use as columns
    aggfunc='sum',                    # Aggregation function
    fill_value=0                       # Fill missing values with 0
)

# Display the pivot table
pop_pivot
```

```python
wb_hdi_by_region = wld_hdi_2014.groupby("Region").agg({"hdi_2014":"mean"}).sort_values("hdi_2014",ascending=False)
wb_hdi_by_region.head()
```

```python
gdp_pivot.columns = gdp_pivot.columns.droplevel(0)
sorted_regions = gdp_pivot.iloc[-1].sort_values(ascending=False).index
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in sorted_regions],
    labels=sorted_regions
)

ax.legend(loc="upper left")
ax.set_title("GDP has grown exponentially over Time")
ax.set_ylabel("GDP (Trillions)")
```

```python
pop_pivot.columns = pop_pivot.columns.droplevel(0)
sorted_regions_pop = pop_pivot.iloc[-1].sort_values(ascending=False).index
```

```python
fig, ax = plt.subplots()

ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in sorted_regions_pop],
    labels=sorted_regions_pop
)

ax.legend(loc="upper left")
ax.set_title("Population has surged from 2 billion to 7.5 billion")
ax.set_ylabel("Population (Billions)")
```

```python
import matplotlib.ticker as mticker

fig, ax = plt.subplots()

minsize = min(wld_hdi_2014["Population (M)"])
maxsize = max(wld_hdi_2014["Population (M)"])
sns.scatterplot(
    data = wld_hdi_2014,
    x = "Life expectancy at birth (years)",
    y = "GDP per capita (USD)",
    size = "Population (M)",
    sizes=(minsize, maxsize),
    hue="Region"

).set(title="Life Expectancy Increases as Countries get richer", yscale="log")

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
wb_hdi_by_region.plot.bar(title="HDI by Region", ylabel = "Human Development Index (HDI)")
```

```python
sns.scatterplot(
    data=wld_hdi_2014.query("`Country Name` != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r"
    
).set(title = "Electricity Drives Development")
```

```python
import matplotlib.gridspec as gridspec
import matplotlib as mpl
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans', 'Bitstream Vera Sans', 'sans-serif']

fig = plt.figure(figsize=(8,12))

gs = gridspec.GridSpec(nrows=12, ncols=6)

stack1_list = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854','#ffd92f', '#b3b3b3']
stack2_list = ['#66c2a5','#a6d854','#b3b3b3','#fc8d62','#e78ac3','#8da0cb','#ffd92f']
bar_list = ['#8da0cb','#fc8d62','#e78ac3','#ffd92f','#66c2a5','#a6d854','#b3b3b3']
buble_list = ['#a6d854','#fc8d62','#ffd92f','#66c2a5','#b3b3b3','#e78ac3','#8da0cb']

fig.suptitle(
    "Report: Global Economic Development from 1960-2018",
    fontsize = 16,
    fontweight="bold",
    x = .65,
    y = .93
)

ax1 = fig.add_subplot(gs[0:4, 0:3])

ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in sorted_regions],
    labels=sorted_regions,
    colors = stack1_list,
    
)
ax1.set_title("GDP has grown exponentially over Time")
ax1.set_ylabel("GDP (Trillions)")

ax2 = fig.add_subplot(gs[0:4, 3:6])

ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in sorted_regions_pop],
    labels=sorted_regions_pop,
    colors = stack2_list
)
ax2.set_title("Population has surged from 2 billion to 7.5 billion")
ax2.set_ylabel("Population (Billions)")
ax2.legend(bbox_to_anchor=(1.75,1), fontsize=8 )

ax3 = fig.add_subplot(gs[4:8, :])

minsize = min(wld_hdi_2014["Population (M)"])
maxsize = max(wld_hdi_2014["Population (M)"])
sns.scatterplot(
    data = wld_hdi_2014,
    x = "Life expectancy at birth (years)",
    y = "GDP per capita (USD)",
    size = "Population (M)",
    sizes=(minsize, maxsize),
    hue="Region",
    palette = buble_list,
    ax=ax3

).set(title="Life Expectancy Increases as Countries get richer", yscale="log")

ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())

handles, labels = (ax3.get_legend_handles_labels())
entries_to_skip = len(wld_hdi_2014["Region"].unique())+1
ax3.legend(handles[entries_to_skip:], labels[entries_to_skip:], bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0., frameon = False, ncol=6)

ax4 = fig.add_subplot(gs[8:12, 0:3])
bars = wb_hdi_by_region.plot.bar(ax=ax4, legend=False)
ax4.set_title("HDI by Region")
ax4.set_ylabel("Human Development Index (HDI)")

for bar, color in zip(bars.patches, bar_list):
    bar.set_color(color)

ax5 = fig.add_subplot(gs[8:12, 3:6])

sns.scatterplot(
    data=wld_hdi_2014.query("`Country Name` != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r",
    ax = ax5
    
).set(title = "Electricity Drives Development")

fig.subplots_adjust(wspace=4, hspace=4)
```
